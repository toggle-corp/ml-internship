## Matrix Product

The equivalent operation for matrices like Dot product for vectors is called the matrix product, or matrix multiplication. It takes two matrices and returns another matrix. This is a core operation in linear algebra.

### Matrices with Vectors

The simpler case of matrix product is between a matrix and a vector (that you can consider as a matrix product with one of them having a single column).

image.png



This illustrates the steps of the product between a matrix and a vector. Letâ€™s consider the first row of the matrix. You do the dot product between the vector (the values 3 and 4 in red) and the row youâ€™re considering (the values 1 and 2 in blue). You multiply the values by pairs: the first value in the row with the first in the column vector (1â‹…3), and the second in the row with the second in the vector (2â‹…4). It gives you the first component of the resulting matrix (1â‹…3+2â‹…4=11).

You can see that the matrix-vector product relates to the dot product. It is like splitting the matrix AA in three rows and applying the dot product.

Let's see how it works in NumPy


```python
import numpy as np
A = np.array([
    [1, 2],
    [5, 6],
    [7, 8]
])
v = np.array([3, 4]).reshape(-1, 1)
print(np.dot(A,v))
```

Note that we used the reshape() function to reshape the vector into a 2 by 1 matrix (the -1 tells Numpy to guess the remaining number). Without it, you would end with a one-dimensional array instead of a two-dimensional array here (a matrix with a single column).

There is another way to think about the matrix product. You can consider that the vector contains **values that weight each column of the matrix**. It clearly shows that the length of the vector needs to be equal to the number of columns of the matrix on which the vector is applied.

### Matrices Product
The matrix product is the equivalent of the dot product operation for two matrices. As youâ€™ll see, it is similar to the matrix-vector product, but applied to each column of the second matrix.
image.png

the resulting matrix has two columns, as the second matrix. The values of first column of the second matrix (3 and 4) weight the two columns and the result fills the first column of the resulting matrix. Similarly, the values of the second column of the second matrix (9 and 0) weight the two columns and the result fills the second column of the resulting matrix.

With Numpy, you can calculate the matrix product exactly as the dot product:


```python
A = np.array([
    [1, 2],
    [5, 6],
    [7, 8],
])
B = np.array([
    [3, 9],
    [4, 0]
])

print(np.dot(A, B))
```

Like with the matrix-vector product, the number of columns of the first matrix must match the number of rows of the second matrix.

## Determinant of a matrix

The determinant of a matrix is the scalar value or number calculated using a square matrix.
It is denoted by:
|A| for a matrix A.


To find a Determinant of a matrix, for every square matrix  [A]nxn there exists a determinant to the matrix such that it represents a unique value given by applying some determinant finding techniques.

image.png

We can calculate the determinants using numpy

```python
A = np.array([[1,1,0], [0,1,0], [0,0,1]])
determinant = np.linalg.det(A)
print ('A = ', A)
print ('determinant of A= ', determinant)
```

### ML alogorithms that use determinants

1. Linear Regressions:
Determinants are not typically used directly in linear regression, as the regression model itself is not based on determinants. However, determinants can be used in some indirect ways to help with linear regression, such as by providing information about the invertibility of a matrix or the rank of a matrix.


- Finding inverse of a matrix

```python
A = np.array([[1,2], [3,4]])
det_A = np.linalg.det(A)

# Check if the matrix is invertible

if det_A != 0:
  A_inv = np.linalg.inv(A)
  print("Inverse of mat A is", A_inv)
else:
  print("Matrix A cannot be inverted")
```

- Calculation of Eigenvalues and eigenvectors:
The eigenvalues and eigenvectors of a matrix are important for understanding the properties of a linear transformation represented by the matrix. In linear regression, the eigenvalues of the design matrix can be used to understand the singularity of the matrix and the condition number of the linear regression solution.

```python
cov_matrix = np.array([[2,1], [1,3]])

# calculating eigen values and eigen vectors

eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

print('Eigen values', eigen_values)
print ('Eigen vectors', eigen_vectors)
```

### Support Vector Machines(SVM):
Determinants are used to calculate the Gram matrix, which is used to determine the optimal hyperplane in SVM algorithms. In SVM, a hyperplane is used to separate the data points into different classes. The optimal hyperplane is chosen such that it maximizes the margin between the classes.
The Gram matrix is a square matrix that represents the dot product of the feature vectors in a feature space.

The Gram matrix can be defined as follows:

G = X * X^T

where X is the feature matrix of the data points and X^T.

The determinant of the Gram matrix is given by:

det(G) = det(X * X^T) = det(X) * det(X^T)

```python
X = np.array([[1,2], [3,4], [5,6]])

# defining a kernel function, here we're using Radial basis function (RBF) kernel
def rbf_kernel(x1, x2, gamma=1):
  return np.exp(gamma * np.linalg.norm(x1-x2)**2)

# calculating the gram matrix using the kernel function
n = X.shape[0]
K = np.zeros((n,n))
for i in range(n):
  for j in range(n):
    K[i,j] = rbf_kernel(X[i], X[j])

print('Gram matrix', K)
```
